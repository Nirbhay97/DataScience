```python
tips = sns.load_dataset('tips')
tips
```

```python
# realtionship between bill and tips
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker',data = tips)
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker',kind ='line',data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker',style = 'sex', size = 'size', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker',style = 'sex', data = tips)
```

```python
sns.catplot(x = 'day', y = 'tip', kind = 'strip', data = tips)
```

```python
sns.catplot(x = 'day', y = 'tip', kind = 'swarm', data = tips)
```

```python
sns.boxplot(tips['tip'])
```

```python
sns.catplot(x = 'day', y = 'tip', kind = 'box', data = tips)
```

```python
sns.catplot(x = 'day', y = 'tip', kind = 'box', hue ='sex', data = tips)
```

```python
sns.violinplot(tips['tip'])
```

```python
sns.catplot(x = 'day', y = 'tip', kind = 'bar', data = tips)
```

```python
sns.catplot(x = 'day', y = 'tip', kind = 'bar', data = tips, estimator = np.median)
```

```python
sns.catplot(x = 'day', y = 'tip', kind = 'bar', data = tips, estimator = np.std)
```

```python
#heatmaps, takes pivot table as input
flights = sns.load_dataset('flights')
flights
```

```python
X = flights.pivot_table(index = 'year', columns = 'month', values = 'passengers', aggfunc = 'sum')
                        
X
```

```python
sns.heatmap(X)
```

```python
sns.heatmap(X, cbar = False)
```

```python
sns.heatmap(X, cbar = False, annot = True)
```

```python
sns.heatmap(X, cbar = False, annot = True, fmt = 'd')
```

```python
sns.heatmap(X, cbar = False, annot = True,cmap = 'winter', fmt = 'd')
```
